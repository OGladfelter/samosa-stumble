```python
import pandas as pd
```

```python
df = pd.read_csv("data/SAMOSA STUMBLE results - SAMOSA STUMBLE IV - 7_12_2025.csv", skiprows=3, header=0)
```

```python
# Keep only the columns we need and rename them
df = df[['Unnamed: 2', 'ACTUAL SAMOSAS EATEN', 'MILES COMPLETED']]
df.columns = ['name', 'samosas', 'miles']

# Clean up any NaN rows
df = df.dropna(subset=['name'])

# Convert to numeric
df['samosas'] = pd.to_numeric(df['samosas'], errors='coerce')
df['miles'] = pd.to_numeric(df['miles'], errors='coerce')

# Save
df.to_csv('data/2025.csv', index=False)
```
